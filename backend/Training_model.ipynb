```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('CarPrice_Assignment.csv')
df.head()
```

Data cleaning

```python
df['st'].unique()
```

```python
df.drop('car_ID',inplace=True,axis = 1)
```

```python
df.drop('CarName',inplace=True,axis = 1)
```

```python
df.dtypes
```

```python
df.drop(columns=['carlength','carwidth','carheight'],inplace= True,axis = 1)
```

```python
df[['wheelbase','boreratio','stroke','compressionratio','price']] = df[['wheelbase','boreratio','stroke','compressionratio','price']].astype(int)
```

```python
df['stroke'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import  make_column_transformer
from sklearn.model_selection import  train_test_split 
from sklearn.pipeline import  make_pipeline 
from sklearn.preprocessing import  StandardScaler
```

```python
x = df.drop('price',axis = 1)
y = df['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 689)
```

```python
from sklearn.linear_model import  LinearRegression
```

```python
lr = LinearRegression()
```

```python
tr = make_column_transformer(
    (OneHotEncoder(sparse_output= False,handle_unknown= 'ignore'),[1,2,3,4,5,6,9,10,12]),
    (StandardScaler(),[0,7,8,11,13,14,15,16,17,18,19]),
    remainder='passthrough'    
)
```

```python
pipe = make_pipeline(tr,lr)
```

```python
pipe
```

```python
pipe.fit(x_train,y_train) 
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score 
```

```python
print(r2_score(y_pred,y_test))
```

```python
# finding best random state  during creating model

score = []
for i in range(1000):
    X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size= 0.2,random_state= i)
    model = LinearRegression()
    pip = make_pipeline(tr,model)
    pip.fit(X_train,Y_train)
    y_pre = pip.predict(X_test)
    score.append(r2_score(Y_test,y_pre))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```
